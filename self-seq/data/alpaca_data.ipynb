```python
alpaca_r_plus_split = [
    "alpaca/alpaca-split_0-c4ai-command-r-plus-GPTQ-generate_instruct-refine-response-final.jsonl",
    "alpaca/alpaca-split_1-c4ai-command-r-plus-GPTQ-generate_instruct-refine-response-final.jsonl",
    "alpaca/alpaca-split_2-c4ai-command-r-plus-GPTQ-generate_instruct-refine-response-final.jsonl",
]
```

```python
# load the data
data = []
for file in alpaca_r_plus_split:
    data.append(pd.read_json(file, lines=True))

df = pd.concat(data, ignore_index=True)
# reindex and save as "idx" column
df['idx'] = df.index
# remove "Input: " from input column and strip with \' \"\
df['input'] = df['input'].str.replace('Input: ', '').str.strip("\'").str.strip("\"")
df['extracted_instruction'] = df['extracted_instruction'].fillna(df['instruction'])

df.head()
```

```python
df.tail()
```

```python
# only keep "input", "system_prompt", "extracted_instruction", "final_instruction_response" columns
df = df[["idx", "input", "system_prompt", "extracted_instruction", "final_instruction_response"]]
df.head()
```

```python
# rename "extracted_instruction" to "instruction", and "final_instruction_response" to "output"
df = df.rename(columns={"extracted_instruction": "instruction", "final_instruction_response": "output"})
df.head()
```

```python
# strip \", \' etc. from instruction
df['instruction'] = df['instruction'].str.strip("\'").str.strip("\"").str.strip("“").str.strip("”")
df.tail()
```

```python
# save the data as jsonl
df.to_json("alpaca/alpaca_r_plus.jsonl", orient="records", lines=True)
```
